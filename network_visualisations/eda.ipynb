```python
# Create a single SQLite database
conn = sqlite3.connect('/Users/tamasmakos/dev/k-monitor/network_analysis/databases/network_analysis.db')

# Load your CSV data into DataFrames
nodes = pd.read_csv('/Users/tamasmakos/dev/k-monitor/network_analysis/node_statistics/node_stats.csv')
subgraphs = pd.read_csv('/Users/tamasmakos/dev/k-monitor/network_analysis/subgraph_statistics/subgraph_stats.csv')
subnetworks = pd.read_csv('/Users/tamasmakos/dev/k-monitor/network_analysis/subnetwork_statistics/subnetwork_stats.csv')

# Correctly naming the tables without the folder path
nodes.to_sql('nodes', conn, if_exists='replace', index=False)
subgraphs.to_sql('subgraphs', conn, if_exists='replace', index=False)
subnetworks.to_sql('subnetworks', conn, if_exists='replace', index=False)
```

```python
# Query the database
query = """
SELECT nodes.*
FROM nodes
JOIN subgraphs ON nodes.Subgraph_ID = subgraphs.Subgraph_ID
WHERE subgraphs.Number_of_Nodes > 30 AND nodes.Winner_or_Buyer = 'Winner'
"""

# Get the results
node_filtered_nodes = pd.read_sql(query, conn)
node_filtered_nodes = node_filtered_nodes.reset_index()
node_filtered_nodes.head()
```

```python
# What is the average degree centrality of the nodes?
mean_degree_centrality = node_filtered_nodes['Number_of_edges'].mean()
print(mean_degree_centrality)
```

```python
# What is the standard deviation of the degree centrality of the nodes?
std_degree_centrality = node_filtered_nodes['Number_of_edges'].std()
print(std_degree_centrality)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from scipy.stats import skew
warnings.filterwarnings('ignore')

# Group by 'Subnetwork_ID', and then within each subnetwork, group by 'Subgraph_ID'.
grouped_subnetworks = node_filtered_nodes[node_filtered_nodes['Winner_or_Buyer'] == 'Winner'].groupby('Subnetwork_ID')

# Calculate the number of subnetworks to determine the number of rows in the subplot
num_subnetworks = len(grouped_subnetworks)
num_cols = 2
num_rows = (num_subnetworks + num_cols - 1) // num_cols

# Create the subplots with a larger figsize
fig, axes = plt.subplots(num_rows, num_cols, figsize=(18, num_rows * 3), squeeze=False)

# Flatten axes array for easy iteration
axes_flat = axes.flatten()

# Define a color palette that can accommodate the maximum number of subgraphs within any subnetwork
max_subgraphs = max(grouped_subnetworks['Subgraph_ID'].nunique())
color_palette = sns.color_palette('husl', n_colors=max_subgraphs)

# Iterate over each subnetwork
for i, (subnetwork, subnetwork_group) in enumerate(grouped_subnetworks):
    ax = axes_flat[i]
    # Group by 'Subgraph_ID' within the subnetwork
    for (subgraph_id, subgraph_group) in subnetwork_group.groupby('Subgraph_ID'):
        # Plot the degree distribution for each subgraph within a subnetwork
        sns.kdeplot(subgraph_group['Number_of_edges'].dropna(), fill=True, ax=ax, label=subgraph_id)
        
        # Calculate skewness
        skewness = skew(subgraph_group['Number_of_edges'].dropna())
        
        # Add skewness as text annotation
        ax.annotate(f'Skewness: {skewness:.2f}', xy=(0.05, 0.9), xycoords='axes fraction')

    ax.axvline(mean_degree_centrality, color='red', linestyle='--', label='Mean Degree Centrality')
    ax.axvline(mean_degree_centrality + std_degree_centrality, color='green', linestyle='--', label='Mean + Std Dev')
    ax.axvline(mean_degree_centrality - std_degree_centrality, color='blue', linestyle='--', label='Mean - Std Dev')
    ax.set_title(f'Degree Distribution for {subnetwork} ({len(subnetwork_group)} nodes)')
    ax.set_xlabel('Number of Edges')
    ax.set_ylabel('Density')
    ax.legend()

# Hide any unused subplots
for j in range(i + 1, len(axes_flat)):
    axes_flat[j].set_visible(False)

# Adjust the layout of the figure
plt.tight_layout()

# Display the figure
plt.show()
```
