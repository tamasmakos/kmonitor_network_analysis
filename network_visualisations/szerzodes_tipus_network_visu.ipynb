```python
import pandas as pd
data = pd.read_csv(r'C:\Users\makos2tamas911\Documents\dev\mm\k_monitor_visu\data_exports\top_15_epitesi_nyero.csv')
data = data[['Eljárás EKR azonosító', 'Ajánlatkérő szervezet neve - tisztított', 'Nyertes ajánlattevő neve - tisztított']]
before = data.shape[0]
data.drop_duplicates(inplace=True)
after = data.shape[0]
print(f'Before: {before}, After: {after}')
data.head(10)
```

```python
data.columns = ['EKR Azonosito', 'Ajánlatkérő', 'Nyertes']
```

```python
import pandas as pd
from pyvis.network import Network
import random

# Assuming 'data' is your pandas DataFrame
# If not, load it first:
# data = pd.read_csv('your_data.csv')

# Create a pyvis network
net = Network(height="750px", width="100%", bgcolor="#FFFFFF", font_color="black", notebook=True, cdn_resources='remote')

# Function to generate a random color
def random_color():
    return "#{:06x}".format(random.randint(0, 0xFFFFFF))

# Add nodes
ajanlatkero_set = set(data['Ajánlatkérő'])
nyertes_set = set(data['Nyertes'])

for ajanlatkero in ajanlatkero_set:
    net.add_node(ajanlatkero, label=ajanlatkero, color='#dd4b39', title=f"Contracting Authority: {ajanlatkero}", x=-400, y=0)

for nyertes in nyertes_set:
    net.add_node(nyertes, label=nyertes, color='#162347', title=f"Winner: {nyertes}", x=400, y=0)

# Add edges
for idx, row in data.iterrows():
    edge_color = random_color()
    net.add_edge(row['Ajánlatkérő'], row['Nyertes'], title=f"Procurement ID: {row['EKR Azonosito']}", color=edge_color)

# Configure other options
net.toggle_physics(True)
net.show_buttons(filter_=['physics'])

# Set the physics layout of the network
net.barnes_hut(gravity=-2000, central_gravity=0.3, spring_length=200, spring_strength=0.05, damping=0.09)

# Try to show the graph
try:
    net.show("procurement_network.html")
except AttributeError:
    print("Couldn't display the graph directly. Saving it to a file instead.")
    net.save_graph("procurement_network.html")
    print("Graph saved to 'procurement_network.html'. Please open this file in a web browser to view the graph.")
```
