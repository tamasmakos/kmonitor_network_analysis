```python
!pip3 install networkx
import networkx as nx

# Provided filter_data function
def filter_data(data, buyer_activity, contract_type, buyer_type, quality_criteria, cost_criteria, price_criteria):
    return data[
        (data['buyer_activity'] == buyer_activity) &
        (data['contract_type'] == contract_type) &
        (data['buyer_type'] == buyer_type) &
        (data['quality_criteria'] == quality_criteria) &
        (data['cost_criteria'] == cost_criteria) &
        (data['price_criteria'] == price_criteria)
    ]

# Function to create a bipartite network from filtered data
# Winners = 0, Buyers = 1
# They are connected if they have a common procurement_id
def create_network(data):
    network = nx.Graph()
    for i, row in data.iterrows():
        # Include the 'place' attribute for each node
        network.add_node(row['winner'], bipartite=0, place=row['place'], bids = row['number_of_bids'])
        network.add_node(row['buyer'], bipartite=1, place=row['place'], bids = row['number_of_bids'])
        network.add_edge(row['winner'], row['buyer'], procurement_id=row['contract_id'])
    return network


# Divide the network data into subnetworks based on categories and store them
subnetworks = {}

for i, row in sorted_data.iterrows():
    filtered_data = filter_data(
        data, row['buyer_activity'], row['contract_type'], row['buyer_type'], row['quality_criteria'], row['cost_criteria'], row['price_criteria']
    )
    subnetwork_name = f"network_{i}"
    subnetworks[subnetwork_name] = {
        'filtered_data': filtered_data,
        'filter_details': {
            'buyer_activity': row['buyer_activity'],
            'contract_type': row['contract_type'],
            'buyer_type': row['buyer_type'],
            'quality_criteria': row['quality_criteria'],
            'cost_criteria': row['cost_criteria'],
            'price_criteria': row['price_criteria']
        },
        'network': create_network(filtered_data)
    }
```

```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import matplotlib.cm as cm
import numpy as np
!pip3 install plotly
import plotly.graph_objects as go


def create_3d_graph_plotly(df, G, buyer_communities_list, winner_communities_list, centrality, filter_details):
    node_x, node_y, node_z, node_color, node_text = [], [], [], [], []
    edge_x, edge_y, edge_z = [], [], []

    # Community colors
    total_communities = len(buyer_communities_list) + len(winner_communities_list)
    community_colors = cm.rainbow(np.linspace(0, 1, total_communities))

    # Assign colors to nodes
    node_color_map = {}
    for idx, comm in enumerate(winner_communities_list):
        color = community_colors[idx]
        for node in comm:
            node_color_map[node] = color

    for idx, comm in enumerate(buyer_communities_list):
        color = community_colors[idx]
        for node in comm:
            node_color_map[node] = 'grey'

    # Unified 3D layout
    pos = nx.spring_layout(G, dim=3, seed=42)
    # Normalize centrality values for node sizes
    size_range = [1, 30]  # Adjust size range as needed
    node_sizes = []
    default_size = 2  # Define a default size for non-winner nodes
    # Assign positions and colors
    for node in G.nodes():
        x, y, z = pos[node]
        node_x.append(x)
        node_y.append(y)
        node_z.append(z)
        color = node_color_map.get(node, 'grey')
        node_color.append(mcolors.to_hex(color))
        node_text.append(f'{node}')
        
        if G.nodes[node]['bipartite'] == 0:  # Winner node
            size = centrality.get(node, 0.5) * (size_range[1] - size_range[0]) + size_range[0]
        else:  # Non-winner node
            size = default_size  # Define a default size for non-winner nodes

        node_sizes.append(size)

    # Assign normalized centrality values to node sizes

    # Create node trace with updated sizes
    node_trace = go.Scatter3d(x=node_x, y=node_y, z=node_z, mode='markers', 
                            marker=dict(color=node_color, size=node_sizes, colorscale='Rainbow'),
                            text=node_text, hoverinfo='text')


    for edge in G.edges():
        x0, y0, z0 = pos[edge[0]]
        x1, y1, z1 = pos[edge[1]]
        edge_x.extend([x0, x1, None])
        edge_y.extend([y0, y1, None])
        edge_z.extend([z0, z1, None])

    edge_trace = go.Scatter3d(x=edge_x, y=edge_y, z=edge_z, mode='lines', 
                              line=dict(color='black', width=1), hoverinfo='none')

    # Create the 3D plot
    fig = go.Figure(data=[edge_trace, node_trace])

    # Format the filter details as a string for annotation
    filter_details_text = "<br>".join([f"{k}: {v}" for k, v in filter_details.items()])

    # Add an annotation for filter details
    fig.add_annotation(
        text=filter_details_text,  # Formatted filter details
        align='left',
        showarrow=False,
        xref='paper',
        yref='paper',
        x=0,  # Position on the left
        y=1,  # Position at the top
        bordercolor='black',
        borderwidth=2,
        bgcolor='white',
        font=dict(
            size=10
        )
    )

    # Update layout with title and legend
    fig.update_layout(
        title='K-MONITOR - EKR SUBNETWORK',
        legend_title_text='Node Type',
        showlegend=False,
        legend=dict(
            itemsizing='constant',
            traceorder='normal',
            font=dict(
                family='sans-serif',
                size=12,
                color='black'
            ),
            bgcolor='LightSteelBlue',
            bordercolor='Black',
            borderwidth=2
        ),
        scene=dict(
            camera=dict(
                up=dict(x=0, y=1, z=0),  # Align 'up' direction with y-axis
                center=dict(x=0, y=0, z=0),
                eye=dict(x=1.25, y=1.25, z=1.25)  # Camera position adjustments
            )
        )
    )

    return fig
```

```python
import plotly.offline as py_offline

# Function to create visualizations and save them
def create_visualization_and_save(df, G, buyer_communities, winner_communities, filter_details, centrality, filename):
    fig = create_3d_graph_plotly(df, G, buyer_communities, winner_communities, centrality, filter_details)
    py_offline.plot(fig, filename=filename, auto_open=False)
```

```python
!pip3 install python-louvain
import community as community_louvain


# Function to detect communities
def detect_communities(G):
    partition = community_louvain.best_partition(G)
    communities = {}
    for node, community_id in partition.items():
        communities.setdefault(community_id, []).append(node)
    return list(communities.values())
```

```python
import networkx as nx

def calculate_community_centrality(G, communities):
    centrality = nx.degree_centrality(G)
    community_centrality = {}
    for community in communities:
        local_max = max([centrality[node] for node in community])
        local_min = min([centrality[node] for node in community])
        for node in community:
            if local_max == local_min:
                normalized_centrality = 0
            else:
                normalized_centrality = (centrality[node] - local_min) / (local_max - local_min)
            community_centrality[node] = normalized_centrality
    return community_centrality
```

```python
def calculate_disparity_measure(subgraph):
    total_degree = sum(dict(subgraph.degree()).values())
    disparity_sums = 0

    for node in subgraph.nodes():
        degree = subgraph.degree(node)
        if degree > 0:
            disparity = sum((subgraph.degree(neighbor) / total_degree) ** 2 for neighbor in subgraph.neighbors(node))
            disparity_sums += disparity

    # Average disparity across all nodes
    average_disparity = disparity_sums / len(subgraph.nodes())
    return average_disparity
```

```python
def project_bipartite(graph, node_type):
    return nx.bipartite.projected_graph(graph, [n for n, d in graph.nodes(data=True) if d['bipartite'] == node_type])
```

```python
import warnings
import os
import pandas as pd
import numpy as np
import networkx as nx

warnings.filterwarnings('ignore')

# Directory for visualizations and centrality measures
output_directory_vis = 'degree_network_visualizations'
output_directory_node_stats = 'node_statistics'
output_directory_subgraph_stats = 'subgraph_statistics'
output_directory_subnetwork_stats = 'subnetwork_statistics'

# Ensure directories exist
os.makedirs(output_directory_vis, exist_ok=True)
os.makedirs(output_directory_node_stats, exist_ok=True)
os.makedirs(output_directory_subgraph_stats, exist_ok=True)
os.makedirs(output_directory_subnetwork_stats, exist_ok=True)

# Initialize DataFrames for various statistics
subnetwork_stats_df = pd.DataFrame()
subgraph_stats_df = pd.DataFrame()
node_stats_df = pd.DataFrame()

# Process each subnetwork
for network_name, subnetwork in subnetworks.items():
    # Extract data and create the network
    network = subnetwork['network']
    filter_details = subnetwork['filter_details']

    # Create and project graphs for buyers and winners
    buyers_graph = nx.bipartite.projected_graph(network, [n for n, d in network.nodes(data=True) if d['bipartite'] == 1])
    winners_graph = nx.bipartite.projected_graph(network, [n for n, d in network.nodes(data=True) if d['bipartite'] == 0])

    # Detect communities and calculate centrality
    buyer_communities = detect_communities(buyers_graph)
    winner_communities = detect_communities(winners_graph)

    # Calculate centrality
    buyer_community_centrality = calculate_community_centrality(buyers_graph, buyer_communities)
    winner_community_centrality = calculate_community_centrality(winners_graph, winner_communities)
    
    # Identify disconnected subgraphs within the subnetwork
    subgraphs = [network.subgraph(c).copy() for c in nx.connected_components(network)]

    # Subnetwork-Level Statistics
    is_connected = nx.is_connected(network)
    communities = None
    if is_connected:
        # Detect communities for modularity calculation (this is an example; choose an appropriate community detection method)
        communities = community_louvain.best_partition(network)

    subnetwork_stats = {
        **filter_details,
        'Subnetwork_ID': network_name,
        'Number_of_Nodes': len(network.nodes()),
        'Number_of_Winners': len([n for n, d in network.nodes(data=True) if d['bipartite'] == 0]),
        'Number_of_Buyers': len([n for n, d in network.nodes(data=True) if d['bipartite'] == 1]),
        'Number_of_Edges': len(network.edges()),
        'Number_of_Subgraphs': len(subgraphs),
        'Assortativity_Coefficient': nx.attribute_assortativity_coefficient(network, 'place'),
        'Density': nx.density(network)
    }
    subnetwork_stats_df = pd.concat([subnetwork_stats_df, pd.DataFrame([subnetwork_stats])], ignore_index=True)
    subnetwork_centrality = {}

    for subgraph_id, sg in enumerate(subgraphs):
        # Check the number of winner nodes in the subgraph
        num_winner_nodes = len([n for n, d in sg.nodes(data=True) if d['bipartite'] == 0])
        if num_winner_nodes < 10:
            continue  # Skip subgraphs with less than 10 winner nodes

        # Calculate centrality measures for each node in the subgraph
        subgraph_centrality = nx.degree(sg)
        
        # Store centrality measures in subnetwork_centrality for winner nodes only
        for node, centrality_value in subgraph_centrality.items():
            if sg.nodes[node]['bipartite'] == 0:  # Check if node is a winner
                subnetwork_centrality[node] = centrality_value

        # Project the subgraph (choose 0 for winners or 1 for buyers)
        projected_sg = project_bipartite(sg, 0)

        community_partition = community_louvain.best_partition(projected_sg)

        # Count unique communities in the partition
        unique_communities = set(community_partition.values())
        # Assign community IDs to nodes in the original subgraph
        for node in sg.nodes():
            projected_node = node  # Assuming the node IDs are consistent between sg and projected_sg
            sg.nodes[node]['community_id'] = community_partition.get(projected_node, None)

        winner_communities_list = detect_communities(projected_sg)
        number_of_cliques = nx.number_of_cliques(sg)
        largest_clique_size = len(max(nx.find_cliques(sg), key=len))
        k_core_number = max(nx.core_number(sg).values())
        is_sg_connected = nx.is_connected(sg)
        disparity_measure = calculate_disparity_measure(sg)
       

        subgraph_stats = {
            **filter_details,
            'Subnetwork_ID': network_name,
            'Subgraph_ID': f"{network_name}_{subgraph_id}",
            'Number_of_Nodes': len(sg.nodes()),
            'Number_of_Edges': len(sg.edges()),
            'Number_of_Winners': len([n for n, d in sg.nodes(data=True) if d['bipartite'] == 0]),
            'Number_of_Buyers': len([n for n, d in sg.nodes(data=True) if d['bipartite'] == 1]),
            'Number_of_Cliques': number_of_cliques,
            'Largest_Clique_Size': largest_clique_size,
            'K_Core_Number': k_core_number,
            'Average_Path_Length': nx.average_shortest_path_length(sg) if is_sg_connected else np.nan,
            'Diameter': nx.diameter(sg) if is_sg_connected else np.nan,
            'Disparity_Measure': disparity_measure,
            'Density': nx.density(sg),
            'Number_of_Winner_Communities': len(unique_communities),
        }
        subgraph_stats_df = pd.concat([subgraph_stats_df, pd.DataFrame([subgraph_stats])], ignore_index=True)

        degree_centrality = nx.degree_centrality(sg)
        betweenness_centrality = nx.betweenness_centrality(sg)
        closeness_centrality = nx.closeness_centrality(sg)
        eigenvector_centrality = nx.eigenvector_centrality(sg, max_iter=10000)

        # Node-Level Analysis
        for node in sg.nodes():
            node_stats = {
                **filter_details,
                'Subnetwork_ID': network_name,
                'Subgraph_ID': f"{network_name}_{subgraph_id}",
                'Node_ID': node,
                'Winner_or_Buyer': 'Winner' if node in winner_community_centrality else 'Buyer',
                'Number_of_edges': sg.degree(node),
                'Community_ID': sg.nodes[node]['community_id'],
                'Degree_Centrality': degree_centrality[node],
                'Betweenness_Centrality': betweenness_centrality[node],
                'Closeness_Centrality': closeness_centrality[node],
                'Eigenvector_Centrality': eigenvector_centrality[node],
                'Clustering_Coefficient': nx.clustering(sg, node),
            }
            node_stats_df = pd.concat([node_stats_df, pd.DataFrame([node_stats])], ignore_index=True)

    # After processing all subgraphs
    filename_vis = os.path.join(output_directory_vis, f"{network_name}_visualization.html")
    create_visualization_and_save(filtered_data, network, buyer_communities, winner_communities, subnetwork['filter_details'], subnetwork_centrality, filename_vis)


# Save all DataFrames to CSV outside the loop
subnetwork_stats_df.to_csv(os.path.join(output_directory_subnetwork_stats, "subnetwork_stats.csv"), index=False)
subgraph_stats_df.to_csv(os.path.join(output_directory_subgraph_stats, "subgraph_stats.csv"), index=False)
node_stats_df.to_csv(os.path.join(output_directory_node_stats, "node_stats.csv"), index=False)
```
